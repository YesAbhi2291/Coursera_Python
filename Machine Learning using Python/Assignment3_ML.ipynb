```python
from sklearn.model_selection import train_test_split

df = pd.read_csv('C:\\Users\\abhir\\OneDrive\\Desktop\\Data Science using Python\\fraud_data.csv')

X = df.iloc[:,:-1]
y = df.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
y
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import recall_score,accuracy_score
dum=DummyClassifier(strategy='most_frequent').fit(X_train,y_train)
y_dummy_predictions=dum.predict(X_test)
acc_sc=accuracy_score(y_test,y_dummy_predictions)
rec_sc=recall_score(y_test,y_dummy_predictions)
print(acc_sc,rec_sc)
```

```python
dum.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import recall_score,accuracy_score,precision_score
sv=SVC(C=1.0).fit(X_train,y_train)
y_predict=sv.predict(X_test)
acc_sc=accuracy_score(y_test,y_predict)
rec_sc=recall_score(y_test,y_predict)
pre_sc=precision_score(y_test,y_predict)
print (acc_sc,rec_sc,pre_sc)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

sv=SVC(C=1e9,gamma=1e-07).fit(X_train,y_train)
y_scores=sv.decision_function(X_test)>-220
clf=confusion_matrix(y_test,y_scores)
clf
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve,auc
import matplotlib.pyplot as plt
clf=LogisticRegression().fit(X_train,y_train)
y_proba=clf.predict_proba(X_test)
y_scores=clf.decision_function(X_test)
precision,recall,thresholds=precision_recall_curve(y_test,y_scores)
#plt.figure()
#plt.plot(precision,recall,label='Precision-Recall plot')
fpr_clf,tpr_clf,_=roc_curve(y_test,y_scores)
roc_auc_clf=auc(fpr_clf,tpr_clf)
#plt.figure()
#plt.plot(fpr_clf,tpr_clf,lw=3,label='AUC is : {:.2f}'.format(roc_auc_clf))
#plt.legend()

#recall[a]
#np.where(fpr_clf==0.16)
#fpr_clf



```
